```python
# Удаление смайликов
def remove_emojis(text):
    if isinstance(text, str):
        emoji_pattern = re.compile(
            "["
            "\U0001F600-\U0001F64F"
            "\U0001F300-\U0001F5FF"
            "\U0001F680-\U0001F6FF"
            "\U0001F1E0-\U0001F1FF"
            "\U00002700-\U000027BF"
            "\U000024C2-\U0001F251"
            "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', text)
    return text

# Парсер
async def main():
    await client.start()

    target_title = "РБК. Новости. Главное"
    dialogs = await client.get_dialogs()

    channel = None
    for dialog in dialogs:
        if dialog.is_channel and dialog.name == target_title:
            channel = dialog
            break

    if channel is None:
        print("Канал не найден")
        return

    messages = []
    async for message in client.iter_messages(channel, limit=200):
        if message is None:
            continue

        has_photo = isinstance(message.media, MessageMediaPhoto)
        has_video = (
            isinstance(message.media, MessageMediaDocument)
            and getattr(message.media.document, "mime_type", "").startswith("video")
        )

        messages.append({
            "date": message.date,
            "text": remove_emojis(message.message),
            "link": f"https://t.me/{channel.entity.username}/{message.id}" if channel.entity.username else None,
            "picture": int(has_photo),
            "video": int(has_video)
        })

    df = pd.DataFrame(messages)
    df["date"] = pd.to_datetime(df["date"]).dt.tz_localize(None)

    # Добавляем колонку с количеством слов
    df["word_count"] = df["text"].apply(lambda x: len(str(x).split()))

    print(f"Скачано сообщений: {len(df)}")
    display(df.head())

    df.to_excel("РБК. Новости. Главное_tg_parsed_no_emoji.xlsx", index=False)
    print("Файл сохранён как: РБК. Новости. Главное_tg_parsed_no_emoji.xlsx")

# Запуск
await main()
```
